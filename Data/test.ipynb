```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("PIB_completo.csv")
data["fecha"] = pd.to_datetime(data["fecha"])
data["pib_mun"] = data["pib_mun"].str.replace(',', '.').astype(float)
data.head()
```

```python
# Calcular la media de cada municipio y ordenarlos de mayor a menor
mean_order = data.groupby("municipio")["pib_mun"].mean().sort_values(ascending=False).index

# Usar el orden deseado en el boxplot
sns.boxplot(x="pib_mun", y="municipio", data=data, order=mean_order)

# Imprimir estadísticas por municipio, ordenadas por media descendente
for municipio in mean_order:
    print(f"Municipio: {municipio}")
    print(data[data["municipio"] == municipio]["pib_mun"].describe())
    print("-" * 40)

plt.show()
```
